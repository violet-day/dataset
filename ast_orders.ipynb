```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('/Users/Nemo/Downloads/backtest_trades_34aa1f643a330171d45ae2896da989f4.csv')
df['Time'] = df['Time'].apply(lambda x:datetime.strptime(x, '%Y-%m-%dT%H:%M:%S%fZ') - timedelta(hours=4))
df['Date'] = df['Time'].apply(lambda x: x.strftime('%Y-%m-%d'))
df['Tag'] = df['Tag ']
df = df[df['Status'].apply(lambda x: x!='Canceled')]
liquidated_symbol = df[df.Tag=='Liquidated ']['Symbol'].unique()

print('trade symbol is %s' % len(df['Symbol'].unique()))
print(liquidated_symbol)

print('liquidated_symbol len is %s' % len(liquidated_symbol))
```

```python
df
```

```python
call_gain = lambda x: (x['Price_sell']-x['Price_buy'])/x['Price_buy']
put_gain = lambda x: (x['Price_sell'] - x['Price_buy'])/x['Price_sell']
```

# put order

```python
sell_df = df[df.Quantity<0][['Symbol', 'Time', 'Price', 'Date']]
buy_df = df[df.Quantity>0][['Symbol', 'Time', 'Price', 'Date']]
order_df = pd.merge(sell_df, buy_df , on=['Symbol', 'Date'], suffixes=('_sell', '_buy'))
order_df['gain'] = order_df.apply(put_gain, axis=1)
order_df
```
